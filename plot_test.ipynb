```python
import pandas as pd
import ast
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
time = pd.read_csv('./time_stats.csv', sep='\t')
freq = pd.read_csv('./freqlen_stats.csv', sep='\t')
```

```python
g0 = sns.FacetGrid(time, col='dataset', hue="model_type",
                  margin_titles=True, height=5, aspect=5/5, 
                  #ylim=(0, 1000),
                  xlim=(0, 35000),
                  palette='tab10'
                 )
g0.map(sns.lineplot, "vocab_size", "train", ci=None, linewidth=3)
g0.add_legend()
g0.savefig('./train_time.png')
```

```python
g = sns.FacetGrid(time, col='dataset', hue="model_type",
                  margin_titles=True, height=5, aspect=5/5, 
                  #ylim=(0, 1000),
                  xlim=(0, 35000),
                  palette='tab10'
                 )
g.map(sns.lineplot, "vocab_size", "token", ci=None, linewidth=3)
g.add_legend()
g.savefig('./token_time.png')
```

```python
g2 = sns.FacetGrid(freq, col='dataset', hue="tokenizer",
                  margin_titles=True, 
                   height=5, 
                   #ylim=(0, 125),
                   xlim=(0, 35000),
                   palette='tab10')
g2.map(sns.lineplot, "vocab_size", "freq@95%", ci=None, linewidth=3)
g2.add_legend()
g2.savefig('./freq@95%.png')
```

```python
g3 = sns.FacetGrid(freq, col='dataset', hue="tokenizer",
                  margin_titles=True,
                   height=5,
                   #ylim=(0, 125),
                   xlim=(0, 35000),
                   palette='tab10'
                  )
g3.map(sns.lineplot, "vocab_size", "avg_len", ci=None, linewidth=3)
g3.add_legend()
g3.savefig('./avg_len.png')
```
