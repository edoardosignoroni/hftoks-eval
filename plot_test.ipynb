```python
import pandas as pd
import ast
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = 'big'

time = pd.read_csv(f'./time_stats_{dataset}.csv', sep='\t')
freq = pd.read_csv(f'./freqlen_stats_{dataset}.csv', sep='\t')
percs = pd.read_csv(f'./en_ga_percs.csv', sep='\t')
```

```python
g0 = sns.FacetGrid(time, col='dataset', hue="model_type",
                  margin_titles=True, height=5, aspect=5/5, 
                  #ylim=(0, 1000),
                  xlim=(0, 35000),
                  palette='tab10'
                 )
g0.map(sns.lineplot, "vocab_size", "train", ci=None, linewidth=3)
g0.add_legend()
g0.savefig(f'./plots/train_time_{dataset}.png')
```

```python
g = sns.FacetGrid(time, col='dataset', hue="model_type",
                  margin_titles=True, height=5, aspect=5/5, 
                  #ylim=(0, 1000),
                  xlim=(0, 35000),
                  palette='tab10'
                 )
g.map(sns.lineplot, "vocab_size", "token", ci=None, linewidth=3)
g.add_legend()
g.savefig(f'./plots/token_time_{dataset}.png')
```

```python
g2 = sns.FacetGrid(freq, col='dataset', row='lang', hue="tokenizer",
                  margin_titles=True, 
                   height=5, 
                   #ylim=(0, 125),
                   xlim=(0, 9000),
                   palette='tab10')
g2.map(sns.lineplot, "vocab_size", "freq@95%", ci=None, linewidth=3)
g2.add_legend()
g2.savefig(f'./plots/freq@95%_{dataset}.png')
```

```python
g3 = sns.FacetGrid(freq, col='dataset', row='lang', hue="tokenizer",
                  margin_titles=True,
                   height=5,
                   #ylim=(0, 125),
                   xlim=(0, 8000),
                   palette='tab10'
                  )
g3.map(sns.lineplot, "vocab_size", "avg_len", ci=None, linewidth=3)
g3.add_legend()
g3.savefig(f'./plots/avg_len_{dataset}.png')
```
