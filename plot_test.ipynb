```python
import pandas as pd
import ast
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#datasets = ['en_ga', 'en_mr', 'en_hi', 'en_lt']
datasets = [ 'ja_my', 'cs_sv', 'syr_zu', 'it_ar']

for dataset in datasets:
    df = pd.read_csv(f'./stats/{dataset}.csv', sep='\t')

    g2 = sns.FacetGrid(df, row='dataset', col='lang', hue="tokenizer",
                      margin_titles=True, 
                       height=5, 
                       #ylim=(0, 280),
                       xlim=(0, 9000),
                       palette='tab10')
    g2.map(sns.lineplot, "vocab_size", "freq@95%", ci=None, linewidth=3)
    g2.add_legend()
    g2.savefig(f'./plots/freq@95_{dataset}.png')
```

```python
#datasets = ['en_ga', 'en_mr', 'en_hi', 'en_lt']
datasets = [ 'ja_my', 'cs_sv', 'syr_zu', 'it_ar']

for dataset in datasets:
    df = pd.read_csv(f'./stats/{dataset}.csv', sep='\t')

    g2 = sns.FacetGrid(df, row='dataset', col='lang', hue="tokenizer",
                      margin_titles=True, 
                       height=5, 
                       ylim=(15, 65),
                       xlim=(0, 8500),
                       palette='tab10')
    g2.map(sns.lineplot, "vocab_size", "avg_len", ci=None, linewidth=3)
    g2.axes[0,0].set_ylabel('μ')
    g2.axes[0,1].set_ylabel('μ')
    g2.axes[0,0].set_xlabel('Vocabulary size')
    g2.axes[0,1].set_xlabel('Vocabulary size')
    
    g2.add_legend()
    g2.savefig(f'./plots/μ_{dataset}.png')
```
