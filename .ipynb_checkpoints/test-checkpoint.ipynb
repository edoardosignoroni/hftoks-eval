```python
import ast
import pandas as pd


df = pd.DataFrame(columns=['model', 'index', 'freq'])

pair = 'ja_my'
lang = 'ja'
size = '4.0'
m1 = 'unigram'
m2 = 'bpe'
m3 = "hft"

def metric(pair, lang, size, model_type):
    
    l1 = pair.split('_')[0]
    l2 = pair.split('_')[1]
    
    f_m = open(f'./tokenizers/{l1}_{l2}/{lang}/{lang}_{model_type}_{size}k/{lang}_{model_type}_{size}k.freq', 'r').read()
    f_m = ast.literal_eval(f_m)
    
    i = 1
    
    weighted_m_val = 0
    #arithmetic_m_val = 0
    m_95_len = int((len(f_m)/100)*95)

    for val in f_m.keys():
        
        weighted_m_val += int(f_m[val])*i
        #arithmetic_m_val += int(f_m[val])
        
        i += 1
        #if i > m_95_len:
        #    break 

    weighted_m_val = weighted_m_val/sum(range(i))
    #arithmetic_m_val = arithmetic_m_val/m_95_len
    
    return weighted_m_val #, arithmetic_m_val

def add_freq_to_df(pair, lang, size, model_type, df):
    
    l1 = pair.split('_')[0]
    l2 = pair.split('_')[1]
    
    f_m = open(f'./tokenizers/{l1}_{l2}/{lang}/{lang}_{model_type}_{size}k/{lang}_{model_type}_{size}k.freq', 'r').read()
    f_m = ast.literal_eval(f_m)
    
    i = 1
    
    for val in f_m.keys():
        if i > 70:   
            df.loc[len(df)+1]=[model_type, i, int(f_m[val])]
        
        i += 1
```

```python
add_freq_to_df(pair, lang, size, m1, df)
add_freq_to_df(pair, lang, size, m2, df)
add_freq_to_df(pair, lang, size, m3, df)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15,15), )
sns.lineplot(data=df, x='index', y='freq', hue='model', linewidth=3)
print(f'{m1} value:', metric(pair, lang, size, m1))
print(f'{m2} value:', metric(pair, lang, size, m2))
print(f'{m3} value:', metric(pair, lang, size, m3))
```
